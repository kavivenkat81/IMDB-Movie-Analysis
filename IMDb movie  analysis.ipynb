```python
# A) Movie Genre Analysis: Analyze the distribution of movie genres and their impact on the IMDB score.
    
#Task: Determine the most common genres of movies in the dataset. 
```

```python
# Splitting genres and extracting the main genre
genres = df['genres'].str.split('|', expand=True)
df['main_genre'] = genres[0]

# Count the occurrences of each genre
genre_counts = df['main_genre'].value_counts()

# Display the most common genres
print(genre_counts)
```

```python
# for each genre, calculate descriptive statistics (mean, median, mode, range, variance, standard deviation) of the IMDB scores.
```

```python
genres = df['genres'].str.split('|', expand=True)
df['main_genre'] = genres[0]  # Assuming the first genre listed is the main genre

# Calculate descriptive statistics for each genre
genre_stats = df.groupby('main_genre')['imdb_score'].agg(['mean', 'median', 'std', 'var', lambda x: x.mode().values[0], lambda x: x.max() - x.min()])
genre_stats.columns = ['Mean', 'Median', 'Std Deviation', 'Variance', 'Mode', 'Range']
print(genre_stats)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Box plot to visualize IMDB scores for different genres
plt.figure(figsize=(12, 6))
sns.boxplot(x='main_genre', y='imdb_score', data=df)
plt.xticks(rotation=45)
plt.title('Distribution of IMDB Scores by Genre')
plt.xlabel('Genre')
plt.ylabel('IMDB Score')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df[df['main_genre'] == 'Action']['imdb_score'], kde=True)
plt.title('IMDB Score Distribution for Action Genre')
plt.xlabel('IMDB Score')
plt.ylabel('Frequency')
plt.show()
```

```python
unique_genres = df['main_genre'].unique()

# Calculate rows and columns required for subplots
num_genres = len(unique_genres)
cols = 4  # Number of columns for subplots
rows = (num_genres + cols - 1) // cols  # Calculate number of rows

# Create subplots for each genre
fig, axes = plt.subplots(rows, cols, figsize=(15, 15))

# Flatten the axes if needed
axes = axes.flatten()

# Iterate through each genre and create histograms
for i, genre in enumerate(unique_genres):
    sns.histplot(df[df['main_genre'] == genre]['imdb_score'], kde=True, ax=axes[i])
    axes[i].set_title(f'IMDB Score Distribution for {genre} Genre')
    axes[i].set_xlabel('IMDB Score')
    axes[i].set_ylabel('Frequency')

# Remove any unused subplots
for j in range(i + 1, rows * cols):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python
# D) Director Analysis: Influence of directors on movie ratings.
# Task: Identify the top directors based on their average IMDB score 
```

```python
director_avg_scores = df.groupby('director_name')['imdb_score'].mean()

# Sorting directors based on average IMDb scores in descending order
top_directors = director_avg_scores.sort_values(ascending=False)

# Displaying the top directors and their average IMDb scores
top_10_director_names = top_directors.head(10).index.tolist()
print(top_10_director_names)
print(top_directors.head(10))
```

```python
# Check for missing values in 'imdb_score' column
missing_imdb_scores = df['imdb_score'].isnull().sum()
print("Number of missing IMDb scores:", missing_imdb_scores)

# Check for unique values in 'director_name' column
unique_directors = df['director_name'].nunique()
total_directors = len(df['director_name'])
print("Number of unique directors:", unique_directors)
print("Total number of directors:", total_directors)
```

```python
import numpy as np

# Grouping directors and their IMDb scores
director_imdb_scores = df.groupby('director_name')['imdb_score'].apply(list)

# Calculate quartiles for each director based on IMDb scores
director_quartiles = director_imdb_scores.apply(lambda x: pd.Series(np.percentile(x, [25, 50, 75]), index=['Q1', 'Q2', 'Q3']))

# Displaying quartiles for directors
print(director_quartiles)
```

```python
# Assuming you have already obtained the top 10 directors' names in the 'top_10_director_names' list

# Filtering the DataFrame for movies directed by the top 10 directors
top_director_movies = df[df['director_name'].isin(top_10_director_names)]

# Grouping the IMDb scores by director name
director_scores_grouped = top_director_movies.groupby('director_name')['imdb_score']

# Calculating percentiles for each director's IMDb scores
director_percentiles = director_scores_grouped.quantile([0.25, 0.5, 0.75]).unstack()

print(director_percentiles)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'director_percentiles' contains the DataFrame with director names and their IMDb score quartiles

# Plotting boxplots for the directors' IMDb score percentiles
plt.figure(figsize=(10, 6))
sns.boxplot(data=director_percentiles, orient='h')
plt.title('IMDb Score Percentiles for Top Directors')
plt.xlabel('IMDb Score')
plt.ylabel('Director Name')
plt.show()
```
